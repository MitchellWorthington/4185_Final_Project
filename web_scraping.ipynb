```python
# first draft of collecting data
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
link = "https://steam250.com/tag/indie"
page = requests.get(link)
page.status_code
```

```python
soup = bs(page.content, "html.parser")
def func(element):
  return element.string
```

```python
# I would include the date but it is of innerHTML and idk how to select that :/
rating = [i for i in range(1, 151)]
title = list(map(func, soup.select("div.appline span.title a"))) # v.col1.main.ranking div#\31 div.appline span.title a
price = list(map(func, soup.select("div.appline > span.price, a.free")))
score = list(map(func, soup.select("div:nth-child(3) > span.score")))
num_votes = list(map(func, soup.select("div:nth-child(3) > span.votes")))
genre = list(map(func, soup.select(" div.appline > a.genre")))
```

```python
data = {
    "Top 150 Rating" : rating,
    "Game Title" : title,
    "Game Price" : price,
    "Score" : score,
    "Votes" : num_votes,
    "Genre" : genre,
}

df = pd.DataFrame(data=data)
```
